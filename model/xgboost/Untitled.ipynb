```python
import numpy as np 
import pandas as pd 
import pickle
```

```python
importance = pd.read_csv('/home/yihang_toby/model/xgboost/feature_importance.csv',index_col= 0)
feat_cols = list(importance.sort_values(by = '0',ascending = False)[:200].index)
```

```python
test = pd.read_csv('/home/yihang_toby/data/train_small.csv')
```

```python
for i in range(len(test)):
    test_df = test.iloc[i:i+1]
```

```python
test_df
```

```python
test_df['date'].item()
```

```python
X = []
for feature in feat_cols:
    if feature.count('feature') == 1:
        pass
        X.append(test_df[feature].item())
    elif feature.count('feature') == 2:
        if '+' in feature:
            feature1,feature2 = feature.split('+')
            X.append(test_df[feature1].item()+test_df[feature2].item())
        elif '-' in feature:
            feature1,feature2 = feature.split('-')
            X.append(test_df[feature1].item()-test_df[feature2].item())
        elif '*' in feature:
            feature1,feature2 = feature.split('*')
            feature2 = feature2[4:-1]
            X.append(test_df[feature1].item() * abs(test_df[feature2].item()))
        else:
            feature1,feature2 = feature.split('/')
            feature2 = feature2[4:-1]
            X.append(test_df[feature1].item()/ abs(test_df[feature2].item() + 1e-5))
    else:
        if '+' in feature:
            feature1 = feature.split('+')[0]
            if '/' in feature:
                feature2,feature3 = feature.split('+')[-1].split('/')
                feature3 = feature3[4:-1]
                X.append((test_df[feature1].item() + test_df[feature2].item()) / (test_df[feature3].item() + 1e-5))
            else:
                feature2,feature3 = feature.split('+')[-1].split('*')
                feature3 = feature3[4:-1]
                X.append((test_df[feature1].item() + test_df[feature2].item()) * (test_df[feature3].item()))
        else:
            feature1 = feature.split('-')[0]
            if '/' in feature:
                feature2,feature3 = feature.split('+')[-1].split('/')
                feature3 = feature3[4:-1]
                X.append((test_df[feature1].item() - test_df[feature2].item()) / (test_df[feature3].item()) + 1e-5)
            else:
                feature2,feature3 = feature.split('+')[-1].split('*')
                feature3 = feature3[4:-1]
                X.append((test_df[feature1].item() - test_df[feature2].item()) * (test_df[feature3].item()))
#         print(feature1,feature2,feature3)
#         method = feature.split('abs')[0][-1]
```
